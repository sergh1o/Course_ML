```python
college = pd.read_csv('Data/college.csv')
college.head()
```

```python
college_isna = college.isna()
college_isnull = college.isnull()
college_isna.equals(college_isnull)
```

```python
ugds = college['ugds']
ugds_operator = ugds + 100
ugds_method = ugds.add(100)
ugds_operator.equals(ugds_method)
```

```python
college_idx = college.set_index('instnm')
sats = college_idx[['satmtmid', 'satvrmid']].dropna()
sats.head()
```

```python
mean = sats.mean()
mean
```

```python
std = sats.std()
std
```

```python
zscore_operator = (sats - mean) / std
zscore_operator.head()
```

```python
zscore_methods = sats.sub(mean).div(std)
zscore_operator.equals(zscore_methods)
```

```python
emp = pd.read_csv('Data/employee.csv')
emp.head()
```

```python
emp.groupby(['DEPARTMENT', 'GENDER']).agg({'BASE_SALARY':'mean'}).round(-3)
```

```python
emp.pivot_table(index='DEPARTMENT', columns='GENDER', 
                values='BASE_SALARY', aggfunc='mean').round(-3)
```

```python
emp.pivot_table(index=['DEPARTMENT','GENDER'], 
                values='BASE_SALARY', aggfunc='mean').round(-3)
```

```python
df1 = emp.groupby('DEPARTMENT').agg({'BASE_SALARY':'mean'}).round(0)
df2 = emp.pivot_table(index='DEPARTMENT', values='BASE_SALARY', 
                      aggfunc='mean').round(0)
df1.equals(df2)
```

```python
emp.pivot_table(index='GENDER', columns='RACE', 
                values='BASE_SALARY', aggfunc='mean').round(-3)
```

```python
pd.crosstab(index=emp['GENDER'], columns=emp['RACE'], 
            values=emp['BASE_SALARY'], aggfunc='mean').round(-3)
```
