```python
# загружаем данные
import pandas as pd
from matplotlib import pyplot
%matplotlib inline
series = pd.read_csv('Data/sunspots.csv', 
                     index_col='Month', 
                     header=0, 
                     squeeze=True)
print(series.head())
series.plot()
pyplot.show()
```

```python
X = series.values
train_size = int(len(X) * 0.66)
train, test = X[0:train_size], X[train_size:len(X)]
print('Общее количество наблюдений: %d' % (len(X)))
print('Количество наблюдений для обучения: %d' % (len(train)))
print('Количество наблюдений для тестирования: %d' % (len(test)))
```

```python
pyplot.plot(train)
pyplot.plot([None for i in train] + [x for x in test])
pyplot.show()
```

```python
from sklearn.model_selection import TimeSeriesSplit
splits = TimeSeriesSplit(n_splits=3)
pyplot.figure(1)
index = 1
for train_index, test_index in splits.split(X):
    train = X[train_index]
    test = X[test_index]
    print('Общее кол-во наблюдений: %d' % (len(train) + len(test)))
    print('Обучающий набор: %d' % (len(train)))
    print('Тестовый набор: %d' % (len(test)))
    pyplot.subplot(310 + index)
    pyplot.plot(train)
    pyplot.plot([None for i in train] + [x for x in test])
    index += 1
pyplot.show()
```

```python
n_train = 500
n_records = len(X)
for i in range(n_train, n_records):
    train, test = X[0:i], X[i:i+1]
    print('train=%d, test=%d' % (len(train), len(test)))
```
